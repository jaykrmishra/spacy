```python
import spacy
from spacy.tokenizer import Tokenizer
from spacy.lang.en import English
```

```python
nlp = spacy.load("en")
text1 = open("test_text.txt","r").read()
```

1. Segmentation

```python
doc1 = nlp(text1)
for sentence in doc1.sents:
    print (sentence)
```

2. Tokenization

```python
#Using spacy 
tokenizer = Tokenizer(nlp.vocab)
tokens = tokenizer(text1)
for token in tokens:
    print (token)
#Using NLTK
#word_tokenize(text2)
#nltk.pos_tag(tokens)
```

3. POS tagging

```python
## Using Spacy

doc = nlp(text1)

for token in doc: 
    print(token, token.pos_) 

print ("Verbs:", [token.text for token in doc if token.pos_ == "VERB"] )

#token.pos and token.pos_ are different
```

4. Name and entity detection

```python
# Using spacy
#Method 1
entities = [(i, i.label, i.label_) for i in doc.ents]
entities
```

```python
#Method 2
for entity in doc.ents:
    print(entity.text + ' - ' + entity.label_ + ' - ' + str(spacy.explain(entity.label_)))
```

5. Lemmatization

```python
for token in doc:
    print (token.text)
    print (token, token.lemma, token.lemma_)
```
